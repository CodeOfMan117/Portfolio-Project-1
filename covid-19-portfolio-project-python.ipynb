# Univariate Analysis

```python
df_corr = df_clean.dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot = True)
```

```python
sns.pairplot(df_corr, vars=['total_cases','people_vaccinated',
                   'total_boosters','total_deaths','handwashing_facilities'],
            hue='gdp_per_capita', palette = 'nipy_spectral')
```

## Investigating Correlation Graphs
* First graph of interest is total_deaths/people vaccinated.There is a linear correlation between the two variables, however countries with a higher GDP/capita seems to have a lower death rate than countries with a lower/negative GDP/capita. This might be due to the facts that the average citizen will have access better healthcare in countries with a higher GDP per capita.
* Another variable that interesting to look at is handwashing_facilities. In all correlation graphs containing the variable, handwashing_facilities, again it is clear that countries with a higher GDP/capita have the means to afford more handwashing_facilities. However it is interesting to note that higher GDP/capita countries, which have more handwashing_facilities, also have a higher number of people vaccinated compared to contries that have a lower GDP/capita. 

```python
sns.pairplot(df_corr, vars=['extreme_poverty','cardiovasc_death_rate',
                   'male_smokers','female_smokers','population_density'],
            hue='gdp_per_capita', palette = 'nipy_spectral')
```

## Investigating Correlation Graphs
* One of the most note worthy observations that can be made from this set of correlation pairplots is that the correlation of both female and male smokers live in extreme_poverty in countries with extremely low GDP/capita compared to countries where the GDP/capita is higher.
* Another interesting correlation is the correlation found between population_density and cadiovasc_death_rate. Interestingly the death rate is higher in countries that have higher population_density, this could be due to the population having to live a more sedintary lifestyle.

```python
sns.pairplot(df_corr, vars=['extreme_poverty','human_development_index',
                   'aged_70_older','aged_65_older','population'],
            hue='gdp_per_capita', palette = 'nipy_spectral')
```

## Investigating Correlation Graphs
* Soo something I found very interesting is the number of individuals, in a population, aged 65 and 70 is higher in countries with lower GDP/Capita compared to countries with higher GDP/Capita.
* Another extemely strong correlation is between population_size and extreme_poverty.
